```python
from Bio.Seq import Seq #importing
dna = Seq("ATGCGCGCTA")
print (dna)
print (type(dna))
```

```python
print(dna.complement()) #Complement function
```

```python
print(dna.reverse_complement()) #DNA is double-stranded and genes can exist on opposite strands.
```

```python
dna = Seq("ATGC")
print(dna.reverse_complement()) #Reverse complement function
```

```python
from Bio.Seq import Seq #Transcribe function to convert a dna to rna
dna = Seq("ATGCGT") #Simply changes T to U
rna = dna.transcribe()
print (rna)
```

```python
from Bio.Seq import Seq #Biopython knows the entire genetic code! No dictionary is needed for this stuff.
dna = Seq("ATGGCCATT")
protein = dna.translate() #Translate function to find codons in a dna sequence and indicate the aminoacids
print (protein)
```

```python
dna = Seq("ATGGCCATTTAA") #TAA is the stop codon and Biopython also knows that!
protein = dna.translate()
print (protein)
```

```python
protein = dna.translate(to_stop = True) #How to make the translation stop as soon as it reaches a stop codon
print (protein)
```

```python
from Bio.SeqUtils import gc_fraction #GC content using the gc_fraction function should first be imported from Bio.SeqUtils
from Bio.Seq import Seq
dna = Seq("ATGCGCGCTA")
gc = gc_fraction (dna)
print (gc)
```

```python
print (gc *100) #to calculate gc content in percents
```

```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
dna = Seq("ATGGCCATTTAA")
print ("Sequence:", dna)
print ("Length:", len(dna))
print ("Complement:", dna.complement())
print ("Reverse Complmenet:", dna.reverse_complement())
rna = dna.transcribe()
print ("RNA:", rna)
protein = dna.translate(to_stop = True)
print ("Protein:", protein)
gc = gc_fraction(dna) * 100
print ("GC Content:", gc)
```
