```python
from Bio import SeqIO
records = SeqIO.parse ("sequence.fasta", "fasta")
for record in records:
    print ("ID:", record.id)
    print ("Sequence:", record.seq)
    print ("Length:", len(record.seq))
```

```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction
records = SeqIO.parse ("sequence.fasta", "fasta")
for record in records:
    gc = gc_fraction (record.seq) * 100
    print ("ID:", record.id)
    print ("GC Content:", gc, "%")
```

```python
from Bio import SeqIO
records = SeqIO.parse ("sequence.fasta", "fasta")
for record in records:
    protein = record.seq.translate(to_stop = True) #.seq.translate, so two functions can be used in a row!
    print ("ID:", record.id)
    print ("Protein:", protein)
#All sequences are not clean and may start improperly or do not translate well. in this example the sequnce were not a multiple of 3 so the codons could not be translated into correct clean amino acids.
```

```python
#Saving results to file
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction
records = SeqIO.parse ("sequence.fasta", "fasta")
output = open("results.txt", "w") #Open a new file named results.txt for writing into it (write mode).
for record in records:
    gc = gc_fraction(record.seq) * 100
    output.write (record.id + "\n") #write into file the record.id and go to the next line. Beware! \n is correct. /n is wrong!
    output.write ("GC:" + str(gc) + "\n\n") #Write into file the GC: and turn gc number into string and go two lines below for next.
output.close() #We have to close the file we have opened.
    
```

```python
#Professional Modern Python style file handling
with open("results2.txt", "w") as output:
    records = SeqIO.parse("sequence.fasta", "fasta")
    for record in records:
        gc = gc_fraction(record.seq) * 100
        output.write ("ID:" + record.id + "\n")
        output.write ("GC:" + str(gc) + "\n\n") #Now there is no need to close the file in this method!
```

```python
with open("results2.txt", "w") as output:
    records = SeqIO.parse("sequence.fasta", "fasta")
    for record in records:
        gc = gc_fraction(record.seq) * 100
        if gc > 50:
            output.write ("ID:" + record.id + "\n")
            output.write ("High GC Content!" + "\n")
            output.write ("GC:" + str(gc) + "\n\n")
        else:
            output.write ("ID:" + record.id + "\n")
            output.write ("GC:" + str(gc) + "\n\n")
```

```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction
high_gc_records = []
records = SeqIO.parse("sequence.fasta", "fasta")
for record in records:
    gc = gc_fraction(record.seq) * 100
    if gc > 50:
        high_gc_records.append(record)
SeqIO.write(high_gc_records, "high_gc.fasta", "fasta") #Creates a .fasta file and writes the high_gc_records data into it.
```
