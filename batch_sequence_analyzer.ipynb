```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction
with open("result.txt", "w") as output:
    records = SeqIO.parse ("Sequence.fasta", "fasta")
    for record in records:
        gc = gc_fraction (record.seq) * 100
        if gc > 50:
            output.write ("ID:" + record.id + "\n")
            output.write ("Sequence:" + str(record.seq) + "\n")
            output.write ("Length:" + str(len(record.seq)) + "\n")
            output.write ("Protein:" + str(record.seq.translate(to_stop = True)) + "\n")
            output.write ("High GC Content!" + "\n")
            output.write ("GC Content:" + str(gc) + "\n\n")
        else:
            output.write ("ID:" + record.id + "\n")
            output.write ("Sequence:" + str(record.seq) + "\n")
            output.write ("Length:" + str(len(record.seq)) + "\n")
            output.write ("Protein:" + str(record.seq.translate(to_stop = True)) + "\n")
            output.write ("GC Content:" + str(gc) + "\n\n")
    
```
